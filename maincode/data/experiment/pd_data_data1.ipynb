```python
belly_data = []
burp_data = []
discomf_data = []
hungry_data = []
tired_data = []
```

```python
belly_folder = 'C:/Users/Desk_Kang/Desktop/Aiffel/workplace/lib/Aiffelthon/data1/belly_pain/'
burp_folder = 'C:/Users/Desk_Kang/Desktop/Aiffel/workplace/lib/Aiffelthon/data1/burping/'
discomf_folder = 'C:/Users/Desk_Kang/Desktop/Aiffel/workplace/lib/Aiffelthon/data1/discomfort/'
hungry_folder = 'C:/Users/Desk_Kang/Desktop/Aiffel/workplace/lib/Aiffelthon/data1/hungry/'
tired_folder = 'C:/Users/Desk_Kang/Desktop/Aiffel/workplace/lib/Aiffelthon/data1/tired/'
```

```python
for filename in os.listdir(belly_folder):
    if filename.endswith(".wav"):
        try:
            # 오디오 파일 로드
            wav, sr = librosa.load(os.path.join(belly_folder, filename), sr=None)
            
            # 오디오의 길이 및 평균값 계산
            duration = librosa.get_duration(y=wav, sr=sr)
            mean_value = wav.mean()
            
            # 결과를 리스트에 추가
            belly_data.append({
                'Filename': filename,
                'Duration (seconds)': duration,
                'Mean Value': mean_value
            })
        except Exception as e:
            print(f"Error processing {filename}: {e}")
```

```python
df_belly_data = pd.DataFrame(belly_data)
```
