```python
variances = {
    'cnn_dist_var': dist_cnn_df.describe().loc['mean', 'row_variance'],
    'cnn_sim_var': sim_cnn_df.describe().loc['mean', 'row_variance'],
    'dist_reg_var': dist_reg_df.describe().loc['mean', 'row_variance'],
    'sim_reg_var': sim_reg_df.describe().loc['mean', 'row_variance'],
    'dtw_var': dtw_df.describe().loc['mean', 'row_variance']
}

rank_lst = []

for value in variances.values():
    rank_lst.append(value)
```

```python
rank_lst = sorted(rank_lst)
```

```python
rank_lst = rank_lst[-1::-1]
rank_lst
```

```python
rank_temp = []

for val in rank_lst:
    for k, v in variances.items():
        if val == v:
            rank_temp.append(k)
rank_temp
```

```python
total = sum(rank_lst)
```

```python
weight = []

for val in rank_lst:
    w = val / total
    weight.append(w)

weight
```

```python
weight_dict = {
    'w_dist_cnn': weight[2],
    'w_sim_cnn': weight[4],
    'w_dist_reg': weight[0],
    'w_sim_reg': weight[1],
    'w_dtw': weight[3],
}
```

```python
total_df = [dist_cnn_df, sim_cnn_df, dist_reg_df, sim_reg_df, dtw_df]

for df in total_df:
    df.drop('row_variance', axis=1, inplace=True)
```

```python
comb_sim_df = pd.concat(total_df, axis=1)
comb_sim_df
```

```python
df['bpain_distcnn_score'] = np.nan
for column in bpain:
    # Determine the weight based on the column suffix
    if column.endswith('dist_cnn'):
        weight_key = 'w_dist_cnn'
    elif column.endswith('sim_cnn'):
        weight_key = 'w_sim_cnn'
    elif column.endswith('dist_reg'):
        weight_key = 'w_dist_reg'
    elif column.endswith('sim_reg'):
        weight_key = 'w_sim_reg'
    elif column.endswith('dtw'):
        weight_key = 'w_dtw'
    else:
        continue  # Skip columns that don't match any condition

    score_v = df[column] * weight_dict[weight_key]

    # Use `isna()` to check for NaN values and update accordingly
    nan_mask = df['bpain_distcnn_score'].isna()
    df.loc[nan_mask, 'bpain_distcnn_score'] = score_v
    df.loc[~nan_mask, 'bpain_distcnn_score'] += score_v
```
